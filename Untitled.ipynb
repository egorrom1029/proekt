```python
import math
import matplotlib.pyplot as plt

cols = num_df.columns
n_cols = 3
n_rows = math.ceil(len(cols) / n_cols)

plt.figure(figsize=(5 * n_cols, 4 * n_rows))

for i, col in enumerate(cols, 1):
    plt.subplot(n_rows, n_cols, i)
    plt.hist(num_df[col].dropna(), bins=15)
    plt.title(col)
    plt.grid(True)

plt.tight_layout()
plt.show()
```

D. Группировки и агрегации

GroupBy + агрегации

```python
grouped = df.groupby('Gender').agg({
    'Sleep Duration': ['mean', 'median'],
    'Age': 'mean',
    'Physical Activity Level': 'mean'
})

grouped
```

Pivot Table

```python
pivot = pd.pivot_table(
    df,
    values='Sleep Duration',
    index='Gender',
    columns='BMI Category',
    aggfunc='mean'
)

pivot
```
